```python
import os
import pandas as pd
import cv2
from insightface.app import FaceAnalysis
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
base_path = './images/'
image_idx = pd.DataFrame( columns=['image_path', 'label_name'])
```

```python
for profile in os.listdir(base_path):
    profile_path = os.path.join(base_path, profile)
    if os.path.isdir(profile_path):
        for image in os.listdir(profile_path):
            image_path = os.path.join(profile_path, image)
            if os.path.isfile(image_path):
                # image_idx = image_idx.append({'image_id': image_path, 'label_name': profile}, ignore_index=True)
                image_idx = pd.concat([image_idx, pd.DataFrame({'image_path': [image_path], 'label_name': [profile]})], ignore_index=True)
```

```python
image_idx
```

```python
recognition = FaceAnalysis(name='buffalo_l')
recognition.prepare(ctx_id=0, det_thresh=0.6, det_size=(640, 640))
```

```python
data = pd.DataFrame(columns = ['Embeddings{}'.format(i) for i in range(512)])
```

```python
for idx, row in image_idx.iterrows():
    image = cv2.imread(row['image_path'])
    image = cv2.resize(image, (600, 600))
    # Detect faces and calculate embeddings using the 'buffalo_l' model
    faces= recognition.get(image)

    for idx, face in enumerate(faces):
        # Calculate the embedding
        face_embedding = face.embedding
        print(type(face_embedding))
        data.loc[len(data)] = face_embedding
```

```python
data
```

```python
result_df = pd.concat([image_idx, data], axis = 1)
result_df
```

```python
result_df.to_csv('data.csv', index = False)
```

```python
required = data.iloc[:, 1:].values

required
```

```python
THRESHOLD = 0.7

def compare_embeddings(new_embedding):
    for embedding in required:
        label = embedding[0]
        embedding = embedding[1:]
        s = cosine_similarity([embedding], [new_embedding])
        if(np.any(s >= THRESHOLD)):
            return label
    return None
```
